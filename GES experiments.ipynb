## Experiment Definition

```python
def G(d):
    G = np.zeros((d, d))
    #G[0,1] = 1.
    return G
    
max_iter = np.inf

adj_CI = None
data_split = False

trials = 25
err_level = .05

loss = 'squared_error' # or 'huber'
clip_data_range = None # or max bound (choose 1 for now please)

eps_max = None
eps_thrsh = None
```

```python
exp = lambda d, n: gaussian_experiment(G(d), n, trials, err_level, adj_CI, data_split,\
                   loss, clip_data_range, eps_max, eps_thrsh, max_iter)  
```

## d=5, usual GES

```python
d = 10
n_inc = 1000
n_iter = 3

ges_results = np.zeros(n_iter)
```

```python
for i in range(n_iter):
    n = (i+1)*n_inc
    ges_results[i] = np.array(gaussian_experiment(G(d), n, trials, err_level, adj_CI, \
                   loss, clip_data_range, eps_max, eps_thrsh, max_iter))  
    print("n=", n, ", metrics=", ges_results[i])
```
